```python
# Imports
from preprocessing import *
import matplotlib.pyplot as plt
```

```python
dicts = process_user_folders("data", "exploration")

# Init lists to append values to
accel = []
left = []
right = []
bullet_fired = []
bullet_missed = []
enemy_killed = []
enemy_collision = []
time = []
damage_taken = []
bullet_distances = []

# Extract data values from dicts
for d in dicts:
    accel.append(d["accel"])
    left.append(d["left"])
    right.append(d["right"])
    bullet_fired.append(d["bullet_fired"])
    bullet_missed.append(d["bullet_missed"])
    enemy_killed.append(d["enemy_killed"])
    enemy_collision.append(d["enemy_collision"])
    time.append(d["time"])
    damage_taken.append(d["damage_taken"])
    bullet_distances.extend(d["bullet_distances"])

# Change None to 0
left = [0 if v is None else v for v in left]
right = [0 if v is None else v for v in right]
bullet_fired = [0 if v is None else v for v in bullet_fired]
bullet_missed = [0 if v is None else v for v in bullet_missed]
enemy_killed = [0 if v is None else v for v in enemy_killed]
enemy_collision = [0 if v is None else v for v in enemy_collision]
```

```python
# Print min and max values
print("accel")
print(min(accel))
print(max(accel))
print("left")
print(min(left))
print(max(left))
print("right")
print(min(right))
print(max(right))
print("bullet_fired")
print(min(bullet_fired))
print(max(bullet_fired))
print("bullet_missed")
print(min(bullet_missed))
print(max(bullet_missed))
print("enemy_killed")
print(min(enemy_killed))
print(max(enemy_killed))
print("enemy_collision")
print(min(enemy_collision))
print(max(enemy_collision))
print("time")
print(min(time))
print(max(time))
print("damage_taken")
print(min(damage_taken))
print(max(damage_taken))
print("bullet_distances")
print(min(bullet_distances))
print(max(bullet_distances))
```

```python
# Create histograms for all the data values
fig, axs = plt.subplots(5, 2, figsize=(10,19))
axs[0, 0].hist(time, bins=20)
axs[0, 0].set_title('Time')
axs[0, 1].hist(accel, bins=20)
axs[0, 1].set_title('Acceleration movements')
axs[1, 0].hist(left, bins=20)
axs[1, 0].set_title('Left movements')
axs[1, 1].hist(right, bins=20)
axs[1, 1].set_title('Right movements')
axs[2, 0].hist(bullet_fired, bins=20)
axs[2, 0].set_title('Bullets fired')
axs[2, 1].hist(bullet_missed, bins=20)
axs[2, 1].set_title('Bullets missed')
axs[3, 0].hist(enemy_killed, bins=20)
axs[3, 0].set_title('Enemies killed')
axs[3, 1].hist(enemy_collision, bins=20)
axs[3, 1].set_title('Enemy collisions')
axs[4, 0].hist(bullet_distances, bins=20)
axs[4, 0].set_title('Bullet distances')
axs[4, 1].hist(damage_taken, bins=20)
axs[4, 1].set_title('Total damage taken')

# Scuffed ax.label_outer() but it keeps tick values
i = 0
for ax in axs.flat:
    ax.yaxis.get_major_locator().set_params(integer=True)
    if i % 2 == 0:
        ax.set(ylabel='Occurrences')
    if i == 8 or i == 9:
        ax.set(xlabel='Values')
    i += 1
```
